````python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
import re
````

````python
def parse_query(text):
    pattern = r"<([^>]+)>(.*?)</\1>"
    matches = re.findall(pattern, text, re.DOTALL)
    return {key: value.strip() for key, value in matches}
````

# Compare Claude few-shots

````python
file_name = 'compare_csv/arc_ai2_chall_train_top_50.csv'
df = pd.read_csv(file_name)
df.head(3)
````

````python
# old = 'claude_v4'

one_v = 'claude_v5'
two_v = 'claude_v6'

for i in range(50):
    row = df.iloc[i]
    if row[one_v] == row[two_v]:
        continue
    else:
        print(i, '\n')
        # old_parsed = parse_query(row[old])
        one_parsed = parse_query(row[one_v])
        two_parsed = parse_query(row[two_v])
        for key in one_parsed:
            if one_parsed[key] != two_parsed[key]:
                print(f'{key}:\n\t{one_parsed[key]}\n\t{two_parsed[key]}\n')
                # print(f'{key}:\n\t{one_parsed[key]}\n\t{two_parsed[key]}\n{old}:\n{old_parsed[key]}\n')
        print('\n---------------------------\n')
````

### Why This Improved Prompt is Better

1.  **Explicit Role-Setting (`SYSTEM` prompt):** Giving the model a persona ("You are an expert translator specializing in educational content...") primes it to adopt the right context and tone from the very beginning. It's a powerful way to improve performance.

2.  **Specific Target Audience:** "Grade school" is broad. Specifying **"Israeli middle school students (grades 7-9)"** is much more precise. This helps the model select the appropriate vocabulary and sentence complexity. Mentioning `חטיבת ביניים` gives it the exact Hebrew context.

3.  **Emphasis on Standard Terminology:** The instruction to use terminology from the **"Israeli science curriculum"** is crucial. This prevents the model from choosing a literally correct but unused term.
    *   For instance, in your original example, "converging of crustal plates" was translated well, but `התנגשות של לוחות טקטוניים` (collision of tectonic plates) is more common and precise terminology in Israeli geology lessons than `התכנסות של לוחות קרום כדור הארץ`. I've updated this in the examples to guide the model better.

4.  **Explicit Grammar Instruction:** Hebrew is a gendered language. Explicitly reminding the model to pay attention to **gender and number agreement** can prevent common, subtle errors that a fast model might otherwise make.

5.  **Clearer Formatting Instructions:** The instruction "You MUST use the following XML-style structure..." is more direct than "Maintain the same formatting." It removes any ambiguity about the expected output.

6.  **Improved Examples:** The examples are slightly tweaked to better reflect the new guidelines (e.g., the tectonic plates example), making the few-shot learning process more effective. Adding a colon (`:`) at the end of a question that leads into a list of options (`...כתוצאה מ:`) is also a common and clear convention in Hebrew tests.

By making these changes, you are providing the model with a much clearer and more constrained set of instructions, which will significantly increase the quality and consistency of your translations for the ARC benchmark.

High-quality prompts with the following features tend to be "upwardly compatible," meaning they work just as well, or even better, with more capable models:

1.  **Clear Role-Playing:** `You are an expert translator...`
2.  **Specific Context:** `...for the ARC_AI2 benchmark... for Israeli middle school students.`
3.  **Explicit Constraints:** `Translate only...,` `Pay close attention to grammar...`

A more advanced model like a hypothetical "Gemini 2.5 Flash" would simply be better at understanding and executing these precise instructions. It would likely make fewer errors in grammar, choose even more appropriate terminology, and adhere to the format more reliably.

You could ask the model to first *reason* about its translation choices before giving the final answer. This forces the model to "think" about the nuances and often leads to a better result.
